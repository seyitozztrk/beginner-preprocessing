#### Visualizing Data

1.Univariate analysis

```python
#**** "Age" distibution using histogram, pdf and cdf (using matplotlib  and seaborn)

#new_df["age"].plot.hist(cumulative=True,bins=1000)
#fig = plt.figure()
#axs = fig.add_subplot(1,1,1)

#matplotlib pdf
plt.hist(df_subset["age"], cumulative=True)


#matplotlib-cdf
x = np.sort(df_subset['age'])
y = np.arange(1, len(x)+1) / len(x)
_ = plt.plot(x,y,marker='.', linestyle='none')
plt.margins(0.02)
plt.show()


```
